```python
def genRandom():
  y = 0
  x = random.random() * 2
  if x < 1:
    y = x
  else:
    y = -x + 2
  return x, y
```

```python
# x, y = genRandom()
# print(f"x = {x} y = {y}")
```

```python
smallXList = []
bigXList = []
smallYList = []
bigYList = []
for i in range(1000):
  x, y = genRandom()
  if x < 1:
    smallXList.append(x)
    smallYList.append(y)
  else:
    bigXList.append(x)
    bigYList.append(y)

plt.hist(smallYList, bins=30, color='blue')
plt.hist(bigYList, bins=30, color='red')
plt.show()
```
