```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nSteps = 30;
nTrials = 200000;
H = [0]*nTrials;
for j in range(nTrials):
  # do all nSteps coin tosses
  T = 2*(np.random.rand(nSteps) < 0.5 ) -1;
  H[j] = sum(T);
plt.hist(H)
plt.show()
```

```python
def makeAstep(sigma):
  theta = 2*np.pi*np.random.random()
  N = np.random.normal(0,sigma)
  deltx = N*np.cos(theta)
  delty = N*np.sin(theta)
  return deltx,delty
```

```python
sigma=2; # step size std dev
nSteps = 2000 # # steps in each walk
x = [0]*nSteps; y = [0]*nSteps
```

```python
for i in range(1,nSteps):
  deltx,delty = makeAstep(sigma)
  x[i]= x[i-1]+deltx; y[i]= y[i-1]+delty
```

```python
#typical walk
plt.subplot(2,2,1) #upper left
plt.plot(x,y)
plt.title(f"(A) Typical walk for {str(nSteps)} steps")
plt.xlabel("X position")
plt.ylabel("Y position")
```

```python
#do the walk from subStart to subEnd
sub1Start = 500;
sub1End = 1500;
xsub1 = [0]*(sub1End-sub1Start)
ysub1 = [0]*(sub1End-sub1Start)
k = 0;
```

```python
for i in range(sub1Start,sub1End):
  xsub1[k] = x[i];
  ysub1[k] = y[i];
  k = k+1;
```

```python
plt.subplot(2,2,2);
plt.title("Walk (A) from step 500 to 1500");
plt.plot(xsub1,ysub1);
```

```python
sub2Start = 1000; sub2End = 1300;
xsub2 = [0]*(sub2End-sub2Start)
ysub2 = [0]*(sub2End-sub2Start)
k = 0;
```

```python
for i in range(sub2Start,sub2End):
  xsub2[k] = x[i];
  ysub2[k] = y[i];
  k = k+1;
```

```python
plt.subplot(2,2,3);
plt.title("Walk (A) from step 1000 to 1300");
plt.plot(xsub2,ysub2);
```

```python
# make a density plot
nWalks = 400;
Endx = [0]*nWalks; Endy = [0]*nWalks;
```

```python
for j in range(nWalks):
  x = [0]*nSteps; y = [0]*nSteps;
  for i in range(1,nSteps):
    deltx,delty = makeAstep(sigma);
    x[i]= x[i-1]+deltx; y[i]= y[i-1]+delty;
  Endx[j] = x[-1];
  Endy[j] = y[-1];
```

```python
#density plot
plt.subplot(2,2,4);
plt.scatter(Endx,Endy,s=3);
plt.title(f"Density Plot for {str(nSteps)} steps");
plt.xlabel("X end");
plt.ylabel("Y end");
plt.tight_layout();
plt.show()
```
